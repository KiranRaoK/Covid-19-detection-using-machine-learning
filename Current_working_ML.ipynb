```python
import pandas as pd
import numpy as np
```

```python
file=open("data.txt",'r')
data=list()
for i in file:
    line=i.split()
    data.append(line)
datawh=data[1:]
df=pd.DataFrame(datawh)
df.columns=['Breathlessness',
 'Sore_throat',
 'Fever',
 'Cough',
 'Dry_Cough',
 'Muscle_pain',
 'Fatigue',
 'Diarrhea',
 'Sneezing',
 'Covid_Presence']
df.head(5)   
  
```

```python
from sklearn import preprocessing as p
LE=p.LabelEncoder()
df['Covid_Presence']=LE.fit_transform(df['Covid_Presence'])
df[['Breathlessness']]=LE.fit_transform(df[['Breathlessness']])
df[['Sore_throat']]=LE.fit_transform(df[['Sore_throat']])
df[['Fever']]=LE.fit_transform(df[['Fever']])
df[[ 'Cough']]=LE.fit_transform(df[[ 'Cough']])
df[['Dry_Cough']]=LE.fit_transform(df[['Dry_Cough']])
df[[ 'Muscle_pain']]=LE.fit_transform(df[[ 'Muscle_pain']])
df[[ 'Fatigue']]=LE.fit_transform(df[[ 'Fatigue']])
df[[ 'Diarrhea']]=LE.fit_transform(df[[ 'Diarrhea']])
df[[ 'Sneezing']]=LE.fit_transform(df[[ 'Sneezing']])
df.head()
```
